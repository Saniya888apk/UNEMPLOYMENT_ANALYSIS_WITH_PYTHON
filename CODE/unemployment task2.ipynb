```python
import pandas as pd


data = pd.read_csv("Unemployment in India.csv")
```

```python

data.columns = data.columns.str.strip()


data.head()


```

```python
# Dataset summary
data.info()

# Missing values check
data.isnull().sum()
```

```python
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=data)
plt.xticks(rotation=45)
plt.title('Unemployment Rate Over Time', fontsize=14)
plt.ylabel('Unemployment Rate (%)')
plt.xlabel('Date')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='Estimated Unemployment Rate (%)', y='Region', data=data)
plt.title('Region-wise Unemployment Rate', fontsize=14)
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Region')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data['Estimated Unemployment Rate (%)'], bins=20, kde=True)
plt.title('Distribution of Unemployment Rate', fontsize=14)
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
